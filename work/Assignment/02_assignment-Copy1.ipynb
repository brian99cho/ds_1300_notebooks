>Using the "state_dates.tsv" data, answer the remaining questions. You will need to merge the two data sets together:

>7) Of the states that joined the United States before 1790, what is the most common state flower?
>8) Which has the larger population density, the most dense US Territory or the least dense state?
>9) Make a graph that plots the populations of the largest city in each entity in the order in which they joined the US. Make the bars black
>10) Make two additional graphs like the one above but one for land area (green bars) and one for water area (blue bars)

Hint: `pd.read_csv('../data/state_dates.tsv',delimiter="\t")`

Hint: You likely want to convert the Date column to datetime. You might have to correct errors in the data as well.

Hint: `states['Date']<pd.datetime(1790,1,1)`

Hint: `pd.merge(****,****,left_on='USPS_code',right_on='Abbreviation',how='outer')`

```python
dates=pd.read_csv('state_dates.tsv',delimiter="\t")
merged = pd.merge(facts, dates,left_on='USPS_code',right_on='Abbreviation',how='outer').drop(columns=['Name'])
merged['Date'] = pd.to_datetime(merged['Date'],errors='raise')


pre_1790=merged[merged['Date']>pd.datetime(1790,1,1)]

most_common_flower = pre_1790['State_flower'].mode()[0]
print(most_common_flower)
```

```python
#8
# Separate territories from states
territories = ['American Samoa', 'Guam', 'Northern Mariana Islands', 'Puerto Rico', 'U.S. Virgin Islands']
merged['Type'] = merged['State'].apply(lambda x: 'Territory' if x in territories else 'State')


# Find the most dense territory and the least dense state
merged['Population_Density'] = merged['Pop_2020'] / merged['Area_land']
most_dense_territory = merged[merged['Type'] == 'Territory']['Population_Density'].max()
least_dense_state = merged[merged['Type'] == 'State']['Population_Density'].min()

# Compare population densities
if most_dense_territory > least_dense_state:
    print('The most dense US Territory has a larger population density.')
else:
    print('The least dense state has a larger population density.')
```

```python
#9
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
plt.bar(merged['State'], merged['city_1_pop'], color='black')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.xlabel('State')
plt.ylabel('Population of Largest City')
plt.title('Population of Largest City in Each State in Joining Order')
plt.show()
```

```python
#10
plt.figure(figsize=(10,8))
plt.bar(merged['State'], merged['Area_land'], color='green')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.xlabel('State')
plt.ylabel('Land Area')
plt.title('Land Area of Each State in Joining Order')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.bar(merged['State'], merged['Area_water'], color='blue')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.xlabel('State')
plt.ylabel('Water Area')
plt.title('Water Area of Each State in Joining Order')
plt.show()
```
